```python
n = int(input("Enter a number: "))
rev = 0
temp = n
num_digits = len(str(n))

for i in range(num_digits):   
    digit = temp % 10
    rev = rev * 10 + digit
    temp //= 10

print("Reversed number is =", rev)
```

```python
num= int(input("enter a number: "))
count = len(str(num))
sum_value= 0

temp=num
while temp >0:
    digit = temp % 10 
    sum_value += digit ** count
    temp //=10
if num == sum_value:
    print("entered number  is arm strong number: ")
else:
    print("entered number is not arm strong number: ")
```
